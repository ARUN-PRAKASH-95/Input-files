```python
from sympy import init_printing,latex,symbols
import sympy as sp
import numpy as np
init_printing()



d1,d2,d3,d4,d5,d6 = sp.symbols('d_{1} d_{2}  d_{3} d_{4} d_{5} d_{6}')
E1,E2,E3 = sp.symbols('E_{1} E_{2}  E_{3}')
G12,G23,G13 = sp.symbols('G_{12} G_{23}  G_{13}')
nu12,nu13,nu21,nu23,nu31,nu32 = sp.symbols('\\nu_{12} \\nu_{13}  \\nu_{21} \\nu_{23} \\nu_{31} \\nu_{32}')

H = sp.Matrix([[1/((1 - d1)*E1),   nu12/E1,   nu13/E1, 0, 0, 0 ],  
              [nu21/E2,      1/((1 - d2)*E2),  nu23/E1, 0, 0, 0 ],
              [nu31/E3,    nu32/E3,   1/((1 - d3)*E3),  0, 0, 0 ],
              [0,0,0,1/((1 - d4)*G12),0,0 ],  
              [0,0,0,0,1/((1 - d5)*G23),0 ],
              [0,0,0,0,0,1/((1 - d6)*G13)]])

C = sp.simplify(H.inv())
```
